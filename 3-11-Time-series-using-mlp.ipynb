
# split a univariate sequence into samples
def split_sequence(sequence, n_steps):
    X, y = list(), list()
    for i in range(len(sequence)):
        
        # find the end of this pattern
        end_ix = i + n_steps
        
        # check if we are beyond the sequence
        if end_ix > len(sequence)-1:
            break
            
        # gather input and output parts of the pattern
        seq_x, seq_y = sequence[i:end_ix], sequence[end_ix]
        X.append(seq_x)
        y.append(seq_y)
        
    return array(X), array(y)



# define input sequence
raw_seq = [10, 20, 30, 40, 50, 60, 70, 80, 90]



# choose a number of time steps
n_steps = 3

# split into samples
X, y = split_sequence(raw_seq, n_steps)

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import datetime
```

```python

data = pd.read_csv('AirPassengers.csv')
data
```

```python
data.Month = pd.to_datetime(data.Month)
```

```python
dataset_ts=data.set_index("Month")

dataset_ts
```

```python
ts1 = pd.concat(
    
    [dataset_ts.shift(3)["#Passengers"],dataset_ts.shift(2)["#Passengers"], 
     dataset_ts.shift(1)["#Passengers"],dataset_ts["#Passengers"]]
    ,axis=1
    )

ts1.columns=[["c1","c2","c3","target"]]

ts1
```

```python
ts2 = ts1.dropna()

ts2
```

```python
X = ts2[["c1","c2","c3"]].to_numpy()
X
```

```python
y= ts2[["target"]].to_numpy()
y
```

```python
y = y.flatten()
y
```

```python

from keras.models import Sequential
from keras.layers import Dense
```

```python

# define model
model = Sequential()

model.add(Dense(10, 
                activation='relu', 
                input_dim=3))

model.add(Dense(1))

model.compile(optimizer='adam', 
              loss='mse')
```

```python
model.layers[0].activation
```

```python

# fit model

model.fit(X, y, 
          epochs=2000, 
          
          verbose=2)
```
